```python
import graspy
import numpy as np
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from graspy.simulations import sbm

n_communities = [50, 50, 50]
p = [[0.5, 0.1, 0.05],
     [0.1, 0.4, 0.15],
     [0.05, 0.15, 0.3],]

np.random.seed(2)
A = sbm(n_communities, p)
from graspy.embed import AdjacencySpectralEmbed

ase = AdjacencySpectralEmbed()
X = ase.fit_transform(A)
import itertools
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.mixture import GaussianMixture
import seaborn as sns

color_iter = {0:'red', 1: 'green', 2: 'blue'}
```

```python
f, axes = plt.subplots(1,1, figsize=(10, 10)) 
gmm = GaussianMixture(n_components=3, covariance_type='full').fit(X)
Y_, means, covariances = gmm.predict(X), gmm.means_, gmm.covariances_
j = 2
k = 1
for i, (mean, covar) in enumerate(zip(means, covariances)):
    v, w = linalg.eigh(covar)
    print(v)
    print(w)
    v = 2. * np.sqrt(2.) * np.sqrt(v)
    u = w[0] / linalg.norm(w[0])
    if not np.any(Y_ == i):
        continue
    sns.scatterplot(X[Y_ == i, k], X[Y_ == i, j], color=color_iter[i],ax=axes )
    angle = np.arctan(u[1] / u[0])
    angle = 180. * angle / np.pi  # convert to degrees
    ell = mpl.patches.Ellipse([mean[1], mean[0]], v[0], v[1], 180. + angle, color=color_iter[i])
    ell.set_clip_box(axes.bbox)
    ell.set_alpha(0.5)
    axes.add_artist(ell)
```
